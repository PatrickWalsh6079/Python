# Reading files exercise
Take the example I demonstrated in the previous lecture (NBA attendance), write your own program to read the data, store in a list, strip the white space, and display your results.

Submit your program code and the results you got in a text file.

The data file NBA-Attendance-2015.zip is attached here. You will need to extract the NBA-Attendance-2015.txt from the zip file.

The columns of data are:

Team Rank, Team Name, Total Attendance, Average Attendance, Percent Capacity

```python
#  Program used:  2-5-1ReadingFiles.ipynb
# This program reads a file and prints the lines and creates a list of items on the line


# open the file for reading (in the same directory as the program)
NBAfile = open('NBA-Attendance-2015.txt', 'r')


# iterate over the lines of the file and count the number of lines
count = 0
NBAlist = []

for line in NBAfile:

    # increment adds one to the count variable
    count += 1

    # strip the newline at the end of the line (and other white space from ends)
    textline = line.strip()

    # split the line on whitespace
    items = textline.split()
    
    # add the list of items to the NBAlist
    NBAlist.append(items)
    
# print the number of teams read
print('Number of teams: ', count)


# print the lines from the list
teams = []
ticket_prices = []
for line in NBAlist:
    team = line[1]
    price = line[-1]
    teams.append(team)
    ticket_prices.append(price)
    print('Line: ', line)

print('Ticket prices of teams:')
count = 0
for price in ticket_prices:
    print('Team: ' + teams[count] + ' ---> $' + price)
    count += 1
    
NBAfile.close()
```
