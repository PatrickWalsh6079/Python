```python
# Each line of this file has the name of a state, mean Verbal 
# SAT score, and mean Math SAT score.

# ['state', 'mean Verbal SAT score', 'mean Math SAT score']

# After reading the data, write code to do the following:

# 1. Determine the state with the highest mean Verbal score 
# and print your results labeled as the ‘State with highest 
# mean Verbal score is: ’.

# 2. Determine the states with mean Math scores greater than 
# 500. Print your results with appropriate labels and include 
# the state name and their math score.

# Submit the code and the output from your program.
```

```python
import csv

file = 'state_satscores_2004.tsv'

with open(file) as csvfile:
    # csv file reader object must be iterated to print results
    reader = csv.reader(csvfile,  dialect='excel', delimiter='\t')
    
    state = ''
    verbal_sat_high = 1
    
    math_sat_500s = []
    
    # loop through contents of reader object
    for line in reader:
        if int(line[1]) > verbal_sat_high:
            state = line[0]
            verbal_sat_high = int(line[1])
        
        if int(line[2]) > 500:
            math_sat_500s.append(line[0])
            math_sat_500s.append(line[2])
            
        print(line)
```

```python
print('State with the highest Verbal SAT score is:')
print(state + ':', verbal_sat_high)
```

```python
print('States with Math SAT scores over 500:')
for state in math_sat_500s:
    print(state)
```
