```python
state_data = {'State':['Alabama','Alaska','Arizona','Arkansas'],
              'PostCode':['AL','AK','AZ','AR'],
              'Area':['52,423','656,424','*','53,182'],
              'Pop':['4,040,587','550,043','3,665,228','2,350,750']}

# Create a dataframe using pd.DataFrame(state_data, columns=['State','PostCode','Area','Pop'].
# Display the table.
# Index the table by 'State'.
# Replace the '*' with '0'.
# Define a function to replace ',' with ''. (remove the commas in Area and Population)
# Use this function to replace the commas in 'Area' and 'Pop'.
```

```python
import pandas as pd

# Create a dataframe using pd.DataFrame(state_data, columns=['State','PostCode','Area','Pop'].
df = pd.DataFrame(state_data, columns=['State','PostCode','Area','Pop'])

# Display the table.
df.head()
```

```python
# Index the table by 'State'.
df = df.set_index('State')
df
```

```python
# Replace the '*' with '0'.
df.loc[df['Area'] == '*', 'Area'] = 0
df
```

```python
# Define a function to replace ',' with ''. (remove the commas in Area and Population)
def comma_remover(df):
    df['Area'] = df['Area'].replace(',','', regex=True)
    df['Pop'] = df['Pop'].replace(',','', regex=True)
    
    return df
```

```python
# Use this function to replace the commas in 'Area' and 'Pop'.
comma_remover(df)
```
